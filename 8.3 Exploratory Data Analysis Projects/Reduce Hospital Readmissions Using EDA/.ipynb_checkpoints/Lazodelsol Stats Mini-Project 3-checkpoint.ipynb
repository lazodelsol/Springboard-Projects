# Hospital Readmissions Data Analysis and Recommendations for Reduction

### Background
In October 2012, the US government's Center for Medicare and Medicaid Services (CMS) began reducing Medicare payments for Inpatient Prospective Payment System hospitals with excess readmissions. Excess readmissions are measured by a ratio, by dividing a hospital’s number of “predicted” 30-day readmissions for heart attack, heart failure, and pneumonia by the number that would be “expected,” based on an average hospital with similar patients. A ratio greater than 1 indicates excess readmissions.

```python
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import bokeh.plotting as bkp
import seaborn as sns
from scipy import stats
from mpl_toolkits.axes_grid1 import make_axes_locatable
import warnings
warnings.filterwarnings('ignore')
np.random.seed(42)
sns.set()
pd.options.mode.chained_assignment = None  # default='warn'
```

```python
# read in readmissions data provided
hospital_read_df = pd.read_csv('data/cms_hospital_readmissions.csv')
```

## Preliminary Analysis

```python
# deal with missing and inconvenient portions of data 
clean_hospital_read_df = hospital_read_df[hospital_read_df['Number of Discharges'] != 'Not Available']
clean_hospital_read_df.loc[:, 'Number of Discharges'] = clean_hospital_read_df['Number of Discharges'].astype(int)
clean_hospital_read_df = clean_hospital_read_df.sort_values('Number of Discharges')
```

```python
# generate a scatterplot for number of discharges vs. excess rate of readmissions
# lists work better with matplotlib scatterplot function
x = [a for a in clean_hospital_read_df['Number of Discharges'][81:-3]]
y = list(clean_hospital_read_df['Excess Readmission Ratio'][81:-3])

fig, ax = plt.subplots(figsize=(8,5))
ax.scatter(x, y,alpha=0.2)

ax.fill_between([0,350], 1.15, 2, facecolor='red', alpha = .15, interpolate=True)
ax.fill_between([800,2500], .5, .95, facecolor='green', alpha = .15, interpolate=True)

ax.set_xlim([0, max(x)])
ax.set_xlabel('Number of discharges', fontsize=12)
ax.set_ylabel('Excess rate of readmissions', fontsize=12)
ax.set_title('Scatterplot of number of discharges vs. excess rate of readmissions', fontsize=14)

ax.grid(True)
fig.tight_layout()
```

### Question B: Provide support for your arguments and your own recommendations with a statistically sound analysis:

Let's start by setting up the hypotheses we'll be testing:
* $H_{0} =$ There is no correlation between the number of discharges and the excess readmission ratio. 
* $H_{a} =$ There is a correlation between the number of discharges and the excess readmission ratio.

```python
# First, let's pull out the same information used in the scatterplot above into a separate dataframe:
df = clean_hospital_read_df.iloc[81:-3]
df.info()
```

```python
# Now let's split up the group into two parts: One where we have the number of discharges < 100
# and one where we have discharges > 1000
df_low = df[df["Number of Discharges"] <= 100]
df_high = df[df["Number of Discharges"] >= 1000]
```

```python
# Now let's do a pearson correlation test on these two groups
r_low, p_low = stats.pearsonr(df_low["Number of Discharges"], df_low["Excess Readmission Ratio"])
r_high, p_high = stats.pearsonr(df_high["Number of Discharges"], df_high["Excess Readmission Ratio"])
print("From our observations where we have less than 100 discharges, our pearson r and p values are: " + str(r_low) + " and " + str(p_low))
print("From our observations where we have more than 1000 discharges, our pearson r and p values are: " + str(r_high) + " and " + str(p_high))
```

As predicted above, we see a statistically significant correlation for low number of discharges (with our p-value being much lower than $\alpha$ = 0.01) while seeing no statistical significance in our data for high number of discharges. For these values, our statisical significance does not differ from practical significance as the calculated values are not close to the threshold values.

Looking at the scatterplot above, we see the advantage and disadvantage of using this type of plot.

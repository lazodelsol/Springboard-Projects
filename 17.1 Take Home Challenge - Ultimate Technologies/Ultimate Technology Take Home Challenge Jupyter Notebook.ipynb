```python
def rollingmean_agg_plot(timeseries, windowsize, figsize, plot_title, set_vlines = False, n = None):
    rolling_mean = timeseries.rolling(window = windowsize, center = True).mean().fillna(method = "bfill")
    plt.figure(figsize = figsize)
    ax = sns.lineplot(x = timeseries.index, y = "Logins", data = timeseries)
    ax = sns.lineplot(x = rolling_mean.index, y = "Logins", data = rolling_mean)
    if set_vlines == True:
        vlines = [timeseries.index[i] for i in range(n, len(timeseries.index), n)]
        ax.vlines(vlines, 0, ax.get_ylim(), linestyles = "dashed", colors = "r")
    ax.set_xlabel("Timestamp")
    ax.set_ylabel("Login Counts")
    ax.set_title(plot_title)
    plt.show()
```

```python
rollingmean_agg_plot(agg15df, windowsize = 100, figsize = (150, 10),
                     plot_title = "Login Counts (15 minute aggregated data) vs Time", set_vlines = True, n = 96)
```

```python
rollingmean_agg_plot(agghourdf, windowsize = 30, figsize = (150, 10),
                     plot_title = "Login Counts (hourly aggregated data) vs Time", set_vlines = True, n = 24)
```

## Making Predictive Model

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from scipy.cluster.hierarchy import linkage, dendrogram
from sklearn.metrics import mean_squared_error, roc_auc_score, mean_absolute_error
from sklearn.preprocessing import Normalizer, normalize
```

```python
X = np.array(ultimatedf.drop("active", axis = 1))
y = np.array(ultimatedf.active)
feature_list = list(ultimatedf.columns.drop("active"))

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .3, random_state = 42)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size = .2, random_state = 42)
```

```python
print('Training Features Shape:', X_train.shape)
print('Training Labels Shape:', y_train.shape)
print('Validation Features Shape:', X_val.shape)
print('Validation Labels Shape:', y_val.shape)
print('Testing Features Shape:', X_test.shape)
print('Testing Labels Shape:', y_test.shape)
```

```python
X_t = np.array(ultimatedf).T
dendro_features = list(ultimatedf.columns)
```

```python
X_norm = normalize(X_t)
mergings = linkage(X_norm, method = "complete")
plt.figure(figsize = (15, 10))
dendrogram(mergings, labels = dendro_features, leaf_rotation = 90)
plt.show()
```

```python
# Define a heatmap function
def halfHeatMap(df, mirror, corrtype):

    # Create Correlation df
    corr = df.corr(method = corrtype)
    # Plot figsize
    fig, ax = plt.subplots(figsize=(15, 15))
    # Generate Color Map
    colormap = sns.diverging_palette(220, 10, as_cmap=True)

    if mirror == True:
        #Generate Heat Map, allow annotations and place floats in map
        sns.heatmap(corr, cmap=colormap, annot=True, fmt=".2f")

    else:
        # Drop self-correlations
        dropSelf = np.zeros_like(corr)
        dropSelf[np.triu_indices_from(dropSelf)] = True# Generate Color Map
        colormap = sns.diverging_palette(220, 10, as_cmap=True)
        # Generate Heat Map, allow annotations and place floats in map
        sns.heatmap(corr, cmap=colormap, annot=True, fmt=".2f", mask=dropSelf)
        
    #show plot
    plt.show()
    
def matches_to_array(matches, feature_list):
    match_list_array=[]
    for match in matches:
        X = np.array(match[feature_list].drop("winPlacePerc", axis = 1))
        y = np.array(match.winPlacePerc)
        match_list_array.append((X,y))
    return match_list_array
```

```python
halfHeatMap(ultimatedf, mirror = False, corrtype = "pearson")
```

```python
# n_estimator tuning
estimator_options = [5, 7, 10, 15, 20]
estimator_scores = []
estimator_mses = []
for estimator in estimator_options:
    model = RandomForestRegressor(n_estimators = estimator, random_state = 42)
    model.fit(X, y)
    y_pred = model.predict(X_val)
    estimator_scores.append(model.score(X_val, y_val))
    estimator_mses.append(mean_squared_error(y_val, y_pred))
```

```python
for i in range(len(estimator_options)):
    print("n_estimator: {}, R^2: {}, MSE: {}".format(estimator_options[i],  
                                                     estimator_scores[i], 
                                                     estimator_mses[i]))
```

```python
sns.lineplot(x = estimator_options, y = estimator_mses)
```

```python
# max_depth tuning, with n_estimator = 15
depth_options = [5, 7, 10, 15, 20]
depth_oobs = []
depth_scores = []
depth_mses = []
for depth in depth_options:
    model = RandomForestRegressor(n_estimators = 15, max_depth = depth, random_state = 42)
    model.fit(X, y)
    y_pred = model.predict(X_val)
    depth_scores.append(model.score(X_val, y_val))
    depth_mses.append(mean_squared_error(y_val, y_pred))
```

```python
for i in range(len(depth_options)):
    print("max_depth: {}, R^2: {}, MSE: {}".format(depth_options[i], depth_scores[i], depth_mses[i]))
```

```python
sns.lineplot(x = depth_options, y = depth_mses)
```

```python
# max_features tuning, with n_estimator = 15 and max_depth = 20
feature_options = [1, 2, 4, 6, 8]
feature_scores = []
feature_mses = []
for feature in feature_options:
    model = RandomForestRegressor(n_estimators = 15, max_depth = 20, max_features= feature, 
                                  random_state = 42)
    model.fit(X, y)
    y_pred = model.predict(X_val)
    feature_scores.append(model.score(X_val, y_val))
    feature_mses.append(mean_squared_error(y_val, y_pred))
```

```python
for i in range(len(feature_options)):
    print("max_feature: {}, R^2: {}, MSE: {}".format(feature_options[i], feature_scores[i], feature_mses[i]))
```

```python
sns.lineplot(x = feature_options, y = feature_mses)
```

```python
jsondf[["mjtheme", "mjtheme_namecode", "mjthemecode"]]
```

```python
# Since the information we need is all located in the "mjtheme_namecode" column, we'll pull that one out specifically 
namecodedf = jsondf["mjtheme_namecode"]
```

Now we'll begin to tackle these two exercises. The first thing I want to do is clean up the data so that all the missing values are filled in. We do this by:
1. Pulling the list of dictionaries from this dataframe into a separate dataframe (we'll call clean_data) that will automatically list each of the original code: name pairings as a separate row.
2. Replacing all the missing values with the word "None".
3. Creating a dictionary with code: name pairings that we can use to apply to fill in the missing values

```python
clean_data = pd.DataFrame()
for row in namecodedf:
    clean_data = clean_data.append(row)

clean_data = clean_data.replace(r'^\s*$', "None", regex=True)

dict_map = {}
for index, row in clean_data.iterrows():
    if row["name"] != "None":
        dict_map[row["code"]] = row["name"]

dict_map
```

Now that we have a mapping dictionary, let's apply this through a function (defined as mapfix) that will create a column with all the correct values in place with respect to the code column.

```python
def mapfix(row):
    return dict_map[row["code"]]

clean_data["name_fixed"] = clean_data.apply(mapfix, axis = 1)

clean_data = clean_data.drop("name", axis = 1)

clean_data["name_fixed"].value_counts()
```

And now we have a completed list of all the different types of projects and their frequency! Let's do the same thing we did as above and do a quick plot of this data to make it more readable.

```python
plt.figure(figsize=(15,5))
clean_data["name_fixed"].value_counts().plot(kind = "bar")
plt.show()
```

Finally, we need to complete the last exercise: Creating a dataframe with all the missing values filled in. The method I chose for this was to split the list of dictionaries on the "," string value, create a new function that will generate a new list for each row containing the correct code: name pairs from the mapping dictionary we made earlier and apply it to generate a new column with the filled in information.

```python
splitdf = jsondf["mjthemecode"].astype(str).str.split(',')

def mapfix2(row):
    list_result = []
    for code in row:
        temp_dict = {}
        temp_dict["code"] = code
        temp_dict["name"] = dict_map[code]
        list_result.append(temp_dict)
    return list_result

jsondf["mjtheme_namecode_fixed"] = splitdf.apply(mapfix2)
jsondf["mjtheme_namecode_fixed"]
```
